```python
%matplotlib inline
import sys 
import numpy as np
import matplotlib.pyplot as plt
sys.path.append("../../")
from pycircos import *
from Bio import SeqIO
```

```python
#Read Genbank, Set locus
gbk = SeqIO.parse("../genbank_data/GCF_000005845.2_ASM584v2_genomic.gbff","genbank")
gcircle = Gcircle()
gcircle.interspace = 0.0
gcircle.read_locus(gbk, bottom=900, height=0, linewidth=0) 
gcircle.set_locus()
locus_names = list(gcircle.locus_dict.keys())
```

```python
#Calc and visualize CDS density
positive_cds_densities = gcircle.calc_feature_density('NC_000913.3', feat_type="CDS", window_size=10000, requirement=lambda x: x.location.strand==1)
negative_cds_densities = gcircle.calc_feature_density('NC_000913.3', feat_type="CDS", window_size=10000, requirement=lambda x: x.location.strand==-1)
gcircle.heatmap('NC_000913.3', positive_cds_densities, bottom=700, height=100, cmap=plt.cm.Reds)
gcircle.heatmap('NC_000913.3', negative_cds_densities, bottom=600, height=100, cmap=plt.cm.Blues)
```

```python
#Visualize link data
link_data = read_linkdata("segdup.txt") #Read link data set. The data should be formated as TWO-LINE FORMAT of Circos. 
for key in link_data.keys():
    gcircle.chord_plot(link_data[key][0], link_data[key][1], bottom=600, color="#F2BE2B", alpha=1.0) 
gcircle.save()
```
